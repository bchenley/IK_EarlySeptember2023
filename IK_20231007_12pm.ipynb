Write a Python program using NumPy to create a 10x10 matrix filled with random integers between 1 and 100. Find the minimum and maximum values of the matrix along with their indices.

```python

np.random.seed(42)

x = ran(1, 101, (10, 10))

x_min, x_max = np.min(x), np.max(x)
x_argmin, x_argmax = np.argmin(x), np.argmax(x)

print(x)
print()
print(x_min, x_argmin)
print()
print(x_max, x_argmax)

unravel_idx_min = np.unravel_index(x_argmin, (10, 10))
unravel_idx_max = np.unravel_index(x_argmax, (10, 10))

print(unravel_idx_min, unravel_idx_max)

np.ravel_multi_index(unravel_idx_max, (10,10))
```
